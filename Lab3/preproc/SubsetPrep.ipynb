# Subsets' preparation

#### Prepare the two dataset that will be studied by the clustering model  

By using the `dataset_features.csv` dataset we can obtain two datasets: `Contracts_feature_A2020.csv` `Contracts_feature_B2020.csv`. The first one contains data for all the contracts that have been activated in 2020 and 2021, the second one in 2018 and 2019.

```python
import pandas as pd
```

Take the `dataset_features.csv` dataset and clean it

```python
inputPath = "<input_path>\\Contracts_feature.csv"
contracts = pd.read_csv(inputPath)
contracts.drop(labels=['Unnamed: 0'], axis=1, inplace=True)
contracts
```

Get the subsets and remove the 'DATA' feature

```python
conBefore2020 = contracts[(contracts.DATA > '2017-12-31')&(contracts.DATA < '2020-01-01')]
conBefore2020.reset_index(inplace=True)
conBefore2020 = conBefore2020.drop(labels=['DATA'], axis=1)
conBefore2020
```

```python
conAfter2020= contracts[contracts.DATA > '2019-12-31']
conAfter2020.reset_index(inplace=True)
conAfter2020 = conAfter2020.drop(labels=['DATA'], axis=1)
conAfter2020
```

Save them as `Contracts_feature_A2020.csv` and `Contracts_feature_B2020.csv`

```python
output_path1 = "<output_path1>\\Contracts_feature_B2020.csv"
output_path2  = "<output_path2>\\Contracts_feature_A2020.csv"
conBefore2020.to_csv(output_path1, index=False)
conAfter2020.to_csv(output_path2, index=False)
print("done")
```
