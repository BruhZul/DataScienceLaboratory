# Feature tranformation of dataset

In this file the dataset will be modified to make it more suitable for Machine Learning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
from sklearn import preprocessing
```

```python
inputPath = "<input path>\\Contracts_dataset.csv"
outputPath = "<output path>\\Contracts_feature.csv"
```

```python
contracts = pd.read_csv(inputPath)
contracts
```

For both GENERE and CITTADINANZA, we can substitute values as 0 and 1.
In particular, male will be 1 and female 0; Italian 1 and foreigner 0

```python
contracts.loc[contracts.CITTADINANZA == 'ITALIANO', 'CITTADINANZA'] = 1
contracts.loc[contracts.CITTADINANZA == 'STRANIERO', 'CITTADINANZA'] = 0
contracts.loc[contracts.GENERE == 'M', 'GENERE'] = 1
contracts.loc[contracts.GENERE == 'F', 'GENERE'] = 0
contracts.head()
```

In the same way we can do the same thing with age: 0 for undefined, 1 for 15-24, 2 for 25-49, 3 for 50-64, 4 for 65 or more

```python
contracts.loc[contracts.ETA == 'NON DEFINITO', 'ETA'] = 0
contracts.loc[contracts.ETA == '15-24', 'ETA'] = 1
contracts.loc[contracts.ETA == '25-49', 'ETA'] = 2
contracts.loc[contracts.ETA == '50-64', 'ETA'] = 3
contracts.loc[contracts.ETA == "65 o piu'", 'ETA'] = 4
contracts.head()
```

We can also consider educational degrees as a ordinal attribute and we can assign values from 0 onward. In order to reduce the complexity, we can group up all the degrees that referes to a tertiary level of education

```python
contracts.TITOLOSTUDIO.unique().tolist()
```

In general we consider 0 for undefined value, 1  for no education degree and so on until 6 for tertiary education and 7 for post-lauream degrees

```python
replaceDict = {"NON DEFINITO":0,
              "NESSUN TITOLO DI STUDIO":1,
              "LICENZA ELEMENTARE":2,
              "LICENZA MEDIA":3,
              "TITOLO DI ISTRUZIONE SECONDARIA SUPERIORE (SCOLASTICA ED EXTRA-SCOLASTICA) CHE NON PERMETTE L'ACCESSO ALL'UNIVERSITÀ ()":4,
              "DIPLOMA DI ISTRUZIONE SECONDARIA SUPERIORE  CHE PERMETTE L'ACCESSO ALL'UNIVERSITA":5,
              "LAUREA - Vecchio o nuovo ordinamento":6,
              "DIPLOMA UNIVERSITARIO":6,
              "DIPLOMA TERZIARIO EXTRA-UNIVERSITARIO":6,
              "DIPLOMA DI SPECIALIZZAZIONE":7,
              "TITOLO DI DOTTORE DI RICERCA":7,
              "TITOLO DI STUDIO POST-LAUREA":7,
              "MASTER UNIVERSITARIO DI PRIMO LIVELLO":7}

contracts.replace({"TITOLOSTUDIO":replaceDict}, inplace=True)
contracts.head()
```

For the MODALITALAVORO attribute, we can also group up the different types of part-time and substitute the values as 0 for undefined, 1 for part-time and 2 for full-time                                   

```python
contracts.MODALITALAVORO.unique().tolist()
```

```python
replaceDict = {"NON DEFINITO":0,
              "TEMPO PARZIALE ORIZZONTALE":1,
              "TEMPO PARZIALE MISTO":1,
              "TEMPO PARZIALE VERTICALE":1,
              "TEMPO PIENO":2}

contracts.replace({"MODALITALAVORO":replaceDict}, inplace=True)
contracts.head()
```

For the last three attributes, we cannot consider them as ordinal values. We can use the OneHotEncoder for PROVINCIAIMPRESA and CONTRATTO. SETTOREECONOMICODETTAGLIO has too many categories so we can use a Binary Encoder to reduce the dimensionality

```python
print("Number of categories for PROVINCIAIMPRESA: ", len(contracts.PROVINCIAIMPRESA.unique().tolist()))
print("Number of categories for CONTRATTO: ", len(contracts.CONTRATTO.unique().tolist()))
print("Number of categories for SETTOREECONOMICODETTAGLIO: ", len(contracts.SETTOREECONOMICODETTAGLIO.unique().tolist()))
```

```python
import category_encoders as ce
enc = ce.BinaryEncoder()

df_col = contracts.SETTOREECONOMICODETTAGLIO.values.reshape(-1,1)
X = enc.fit_transform(df_col)
contracts = pd.concat([contracts, X], axis=1)

contracts
```

```python
#laber encoder
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
labels = le.fit_transform(contracts['SETTOREECONOMICODETTAGLIO'])
contracts['SETTOREECONOMICODETTAGLIO'] = list(labels)
contracts.head()
```

```python
contracts = pd.get_dummies(contracts, columns=['CONTRATTO'], prefix=['CONTRATTO'], drop_first=True)
contracts.head()
```

```python
contracts = pd.get_dummies(contracts, columns=['PROVINCIAIMPRESA'], prefix=['PROVINCIA'], drop_first=True)
contracts.head()
```

```python
contracts.drop(labels=['SETTOREECONOMICODETTAGLIO'], axis=1, inplace=True)
contracts
```

```python
contracts.to_csv(outputPath, index=True)
print("done")
```
